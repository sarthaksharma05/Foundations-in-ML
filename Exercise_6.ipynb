```python
import numpy as np

# Marks array (5 students Ã— 3 subjects)
marks = np.array([
    [78, 85, 90],
    [45, 60, 55],
    [88, 92, 95],
    [30, 40, 45],
    [65, 70, 75]
])

print(marks)
```

```python
print(marks.shape)
```

```python
print(marks.size)
```

```python
print(marks[1])
```

```python
print(marks[:, 2])
```

```python
print(marks[0, 1])
```

```python
student_avg = np.mean(marks, axis=1)
subject_avg = np.mean(marks, axis=0)

print("Average marks of each student:", student_avg)
print("Average marks of each subject:", subject_avg)
```

```python
student_avg = np.mean(marks, axis=1)
top_students = np.where(student_avg >= 80)

print("Students with average marks >= 80:", top_students[0] + 1)
```

```python
status = np.where(marks >= 50, 1, 0)
print(status)
```

```python
marks[marks < 40] = 0
print(marks)
```
