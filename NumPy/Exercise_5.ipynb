```python
daily_sales = np.array([
    [10, 15, 8, 12],   # Headphones
    [5, 7, 4, 6],      # Smartwatches
    [2, 3, 2, 4]       # Speakers
])
```

```python
total_items = np.sum(daily_sales)
print("Total items sold:", total_items)
```

```python
product_wise_total = np.sum(daily_sales, axis=1)
print("Items sold per product:", product_wise_total)
```
