```python
import pandas as pd
import numpy as np

data = {
    "name": ["Alice", "Bob", "Charlie", None, "Eva", "Frank", "Grace", "Henry", None, "Ivy"],
    "department": ["IT", "HR", "IT", "Marketing", "Finance", "IT", "HR", "Marketing", "Finance", "IT"],
    "role": ["Developer", "Recruiter", "Data Scientist", "Content Manager", "Accountant", "Backend Developer", "HR Manager", "SEO Specialist", "Financial Analyst", "Frontend Developer"],
    "experience": [3, 5, 2, 4, 6, 3, 7, 2, 5, 1],
    "salary": [70000, 60000, 75000, 55000, 80000, 72000, 65000, 50000, 78000, 68000],
    "score": [85, 90, 88, 92, np.nan, 87, 91, 84, 89, np.nan]
}

df = pd.DataFrame(data)

print(df)
```

```python
df.groupby("department")["salary"].mean()
```

```python
df.groupby("role")["score"].mean()
```

```python
df.groupby("department")["score"].count()
```

```python
df.groupby("department").size()
```

```python
df.groupby("department")["salary"].agg(["mean","max","min","count"])
```

```python
df.groupby("role")["score"].agg(["mean","max","min","count"])
```

```python
df.groupby(["department","role"])["salary"].mean()
```

```python
df.groupby(["department","experience"])["score"].mean()
```
