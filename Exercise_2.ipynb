```python
import numpy as np

# Daily sales data
daily_sales = np.array([
    [10, 15, 8, 12],   # Headphones
    [5, 7, 4, 6],      # Smartwatch
    [2, 3, 2, 4]       # Speakers
])

print(daily_sales)

```

```python
bonus_sales = daily_sales + 2
print("Sales after bonus:\n", bonus_sales)

```

```python
# Cost price of each product
cost_price = np.array([1000, 2000, 1500])

# Multiply sales with cost price
revenue = daily_sales * cost_price.reshape(3, 1)

print("Revenue matrix:\n", revenue)

# Total revenue per day
total_revenue = revenue.sum(axis=0)
print("Total revenue per day:\n", total_revenue)
```

```python
sqrt_sales = np.sqrt(daily_sales)
print("Square root of sales:\n", sqrt_sales)
```
