1D Slicing & Indexing 
This code shows different ways to access elements in an array:

Array: [10, 20, 30, 40] (at positions 0, 1, 2, 3)

print(arr[0]) → 10

Gets the first element (position 0)
In programming, counting starts at 0, not 1

print(arr[-1]) → 40
Gets the last element using negative indexing
-1 means "the last one", -2 means "second to last", etc.
print(arr[1:3]) → [20, 30]

Slicing: gets elements from position 1 to 2 (not including 3)
Format: [start:stop] (stop is exclusive)
print(arr[:2]) → [10, 20]

Gets elements from the beginning up to position 2 (not including 2)
Empty before : means "start from the beginning"

print(arr[::2]) → [10, 30]
Gets every 2nd element starting from the beginning
Format: [start:stop:step]
Empty start and stop means "all elements", 2 means "every 2nd"
Think of it like picking items from a list:

arr[0] = "give me item at spot 0"
arr[-1] = "give me the last item"
arr[1:3] = "give me items from spot 1 to 2"
arr[:2] = "give me items from the start up to spot 2"
arr[::2] = "give me every other item"

```python
arr = np.array([10,20,30,40])
print(arr[0]) #Indexing
print(arr[-1])

print(arr[1:3])
print(arr[:2])
print(arr[::2])
```

2D Indexing

This line gets an entire column from the 2D array.

Breaking it down:

mat[:, 1]
: = "all rows"

Where is NumPy

```python
arr = np.array([10,-5,20,-3,15])
result = np.where(arr<0,0,arr)
print(result)
```

```python
arr = np.array([1,2,3,4,5])
result = np.where(arr%2 != 0, -1, arr)
print(result)
```

```python
marks  = np.array([45,78,32,90,55])
status = np.where(marks>=50 , 1, 0)
print(status)
```

```python
data = np.array([10,20,30,40,50])
mean_val = np.mean(data)
result = np.where(data>mean_val , mean_val, data)
print(result)
```
